### Example
Plotting the stock prices of the tech companies. 

````python
google = pd.read_csv(get_path('GOOGL_data.csv'))
fb     = pd.read_csv(get_path('FB_data.csv'))
apple = pd.read_csv(get_path('AAPL_data.csv'))
amazon = pd.read_csv(get_path('AMZN_data.csv'))
microsoft = pd.read_csv(get_path('MSFT_data.csv'))
google.head()
````

### **Bar Graph**

````python
movie_scores = pd.read_csv(get_path('movie_scores.csv'))
movie_scores
````

````python
movies_fig = plt.figure(figsize=(10, 5), dpi=75)
ax = movies_fig.add_subplot(1,1,1)
x = np.arange(1,len(movie_scores.Tomatometer)+1)
width = 0.2
ax.bar(height= movie_scores.Tomatometer , width=width,  x = x-0.3/2)
ax.bar(height =movie_scores.AudienceScore,width=width, x = x+0.3/2)
ax.set_xticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_xticks(x)
````

### For horizontal bar just flip (x,y), (Width, Height) and so on. 

````python
movies_fig = plt.figure(figsize=(10, 5), dpi=75)
ax = movies_fig.add_subplot(1,1,1)
y = np.arange(1,len(movie_scores.Tomatometer)+1)
width = 0.2
ax.barh(width= movie_scores.Tomatometer , height=width,  y = y-0.3/2)
ax.barh(width =movie_scores.AudienceScore,height=width, y = y+0.3/2)
ax.set_yticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_yticks(y)
````

### **Histogram**

The difference between Histogram and barplot is that barplot comparing many columns/attributes together.  
On the other side, Histogram shows the distribution of values for a specific column.

````python
hist_fig, ax = plt.subplots(1,1)
ax.hist(np.random.randn(100),bins=50) #(np.random.randn(100))creates an array of specified shape and fills it with random values as per standard normal distribution
ax.hist(np.random.randn(100),bins=50)
````

````python
df_titanic = pd.read_csv(get_path('titanic.csv'),index_col=0)
df_titanic.head()
````

````python
hist_fig, ax = plt.subplots(1,1)
max_age = max(df_titanic.Age.values)
min_age = min(df_titanic.Age.values)
age_bins= int((max_age-min_age)//10)
ax.hist(df_titanic.Age.values,bins=age_bins,color='k')
ax.grid()
````

### ***Bubble Plot***

Bubble plot is just a scatter plot with the ability to add another 2 dimensions which can be viewed as the scale s and the color c

````python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
sp = ax.scatter(np.random.randn(20),np.random.randn(20),s=np.abs(np.random.randn(20))*100,alpha=0.5,c= np.abs([np.random.randn(20)]))
fig.colorbar(sp)
````

For the titanic data, plot a scattered plot with the *Age* on the x-axis and the *Survival* count on the y-axis, such that the size of the circle is equivalent to the mean *Fare* of the ticket for the passengers of a certain age.
