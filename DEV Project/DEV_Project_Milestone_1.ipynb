Any value after 2370000 is invalid

- 4.5329 PLN = 1 EUR, during March 2021.
- 4.5885 PLN = 1 EUR , during April 2021
- 4.52488 PLN = 1 EUR, during May 2021


> We will use these values to convert offers made in EUR to PLN



```python
print(df.shape)
Curr_rate = pd.DataFrame(
    {'Currency':['PLN', 'PLN', 'PLN' , 'EUR', 'EUR', 'EUR'],
     'year_month':['2021-03','2021-04','2021-05','2021-03','2021-04','2021-05'],
     'To_PLN':[1, 1, 1, 4.5329, 4.5885, 4.52488 ] })

df =pd.merge(df, Curr_rate, left_on=['Currency', 'year_month'], right_on=['Currency', 'year_month'], how='left')
df['Price'] = round((df['Price'] * df['To_PLN']), 2)

print(df.shape)
```
